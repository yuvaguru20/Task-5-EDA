```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading data
df = pd.read_csv("train.csv")

# Show first 5 rows
df.head()
```

```python
# Checking general info about dataset
df.info()

# Summary statistics for numerical columns
df.describe()
```

```python
# Value counts for some categorical columns
print(df['Sex'].value_counts())
print(df['Embarked'].value_counts())
```

```python
sns.pairplot(df[['Survived','Age','Fare','Pclass']], hue='Survived')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['Survived','Pclass','Age','SibSp','Parch','Fare']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df[['Age','Fare']].hist(bins=20, figsize=(10,4))
plt.suptitle("Age and Fare Distribution")
plt.show()


df[['Age','Fare']].hist(bins=20, figsize=(10,4))
plt.suptitle("Age and Fare Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Fare'])
plt.title("Boxplot of Fare")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title("Age vs Fare (colored by Survival)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title("Survival Rate by Sex")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title("Survival Rate by Passenger Class")
plt.show()
```
